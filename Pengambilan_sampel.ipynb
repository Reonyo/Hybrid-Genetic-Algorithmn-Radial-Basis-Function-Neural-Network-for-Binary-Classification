```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'D:\Study\Kuliah Semester 5\Simulasi A2\Tugas Akhir\Deteksi Fraud.csv')
```

```python
# Check Distribution
fraud_distribution = df['fraud_bool'].value_counts()
print(fraud_distribution)
```

```python
fraud_cases = df[df['fraud_bool'] == 1]
non_fraud_cases = df[df['fraud_bool'] == 0]

# 500 Sample for Fraud and 500 Sample for Non-Fraud
fraud_sample = fraud_cases.sample(n=500, random_state=42)
non_fraud_sample = non_fraud_cases.sample(n=500, random_state=42)

# Combining
balanced_df = pd.concat([fraud_sample, non_fraud_sample])
balanced_df = balanced_df.sample(frac=1, random_state=42).reset_index(drop=True)

# Check
print(balanced_df['fraud_bool'].value_counts())
```

```python
columns_to_use = [
    'income',
    'credit_risk_score',
    'customer_age',
    'payment_type',
    'employment_status',
    'current_address_months_count',
    'fraud_bool'  # Target variable
]
# Create a new DataFrame with only the specified columns
df_resample = balanced_df[columns_to_use]
```
